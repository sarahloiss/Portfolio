```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read the dataset
data = pd.read_csv("world_population_data.csv")

data.head()
```

```python

import squarify


# Extract necessary columns
tree_data = data[['country', '2023 population']]


# Plotting the tree map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=tree_data['2023 population'], label=tree_data['country'], alpha=0.8)
plt.title('World Population by Country 2023')
plt.axis('off')  
plt.show()
```

```python
# Area chart of India population



# Selecting only India for the area chart
selected_country = data[data['country'] == 'India']

# Creating a new DataFrame for the area chart
years = ['2023', '2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
india_population_data = pd.DataFrame()

# Extracting population data for each year for India
for year in years:
    year_column = f'{year} population'
    temp_df = selected_country[['country', year_column]]
    temp_df.rename(columns={year_column: 'population'}, inplace=True)
    temp_df['year'] = int(year)
    india_population_data = pd.concat([india_population_data, temp_df])

# Pivot the data for the area chart
india_population_data_pivot = india_population_data.pivot(index='year', columns='country', values='population')

# Plotting the area chart
plt.figure(figsize=(12, 8))
india_population_data_pivot.plot(kind='area', stacked=False)
plt.title('Population of India Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# Stacked Area Chart 


# Reshaping the dataset for a stacked area chart
# Selecting a subset of countries for demonstration. Here, I choose the first 5 countries for simplicity.
selected_countries = data['country'].head(5)

# Creating a new DataFrame for the stacked area chart
years = ['2023', '2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
area_chart_data = pd.DataFrame()

# Extracting population data for each year for the selected countries
for year in years:
    year_column = f'{year} population'
    temp_df = data[data['country'].isin(selected_countries)][['country', year_column]]
    temp_df.rename(columns={year_column: 'population'}, inplace=True)
    temp_df['year'] = int(year)
    area_chart_data = pd.concat([area_chart_data, temp_df])

# Pivot the data for the stacked area chart
area_chart_data_pivot = area_chart_data.pivot(index='year', columns='country', values='population')

# Plotting the stacked area chart
plt.figure(figsize=(12, 8))
area_chart_data_pivot.plot(kind='area', stacked=True)
plt.title('Population Over Time for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
