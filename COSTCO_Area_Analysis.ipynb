```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read the dataset
store_data = pd.read_csv("costcos-geocoded.csv")

store_data.head()
```

```python
import numpy as np
import seaborn as sns

# Heat map of locations
plt.figure(figsize=(10, 6))
sns.kdeplot(store_data['Longitude'], store_data['Latitude'], cmap="Reds", shade=True, bw_adjust=.5)
plt.title('Density Plot of Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python

# Scatter plot for spatial visualization
plt.figure(figsize=(10, 6))
plt.scatter(store_data['Longitude'], store_data['Latitude'], alpha=0.5, c='blue')
plt.title('Spatial Distribution of Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# Preparing data for the lollipop chart by counting the number of locations per state
locations_per_state = store_data['State'].value_counts().reset_index()
locations_per_state.columns = ['State', 'Number of Locations']
locations_per_state = locations_per_state.sort_values(by='Number of Locations', ascending=True)

# Plotting the lollipop chart
plt.figure(figsize=(10, 8))
plt.stem(locations_per_state['State'], locations_per_state['Number of Locations'], basefmt=" ")
plt.xticks(rotation=90)
plt.title('Number of Locations per State')
plt.ylabel('Number of Locations')
plt.xlabel('State')
plt.tight_layout()
plt.show()
```
