```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Read the dataset
crime_data = pd.read_csv("crimerates_by_state_2005.csv")

crime_data.head()
```

```python
# Scatterplot 

# Filtering out the United States total to focus on state-level data
state_data = crime_data[crime_data['state'] != 'United States']

# Scatterplot: Robbery vs. Aggravated Assault
plt.figure(figsize=(10, 6))
sns.scatterplot(x='robbery', y='aggravated_assault', data=state_data)
plt.title('Robbery vs. Aggravated Assault Rates by State in 2005')
plt.xlabel('Robbery Rate (per 100,000 population)')
plt.ylabel('Aggravated Assault Rate (per 100,000 population)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Bubble Chart

sns.scatterplot(x='robbery', y='aggravated_assault', size='population', sizes=(50, 500), alpha=0.5, data=state_data)
plt.title('Robbery vs. Aggravated Assault Rates by State in 2005 (Bubble Size: Population)')
plt.xlabel('Robbery Rate (per 100,000 population)')
plt.ylabel('Aggravated Assault Rate (per 100,000 population)')
plt.legend(title='Population Size', labelspacing=1, loc='upper left')
plt.show()
```

```python
# Density Plot

plt.figure(figsize=(10, 6))
sns.kdeplot(state_data['murder'], shade=True, color="r")
plt.title('Density Plot of Murder Rates by State in 2005')
plt.xlabel('Murder Rate (per 100,000 population)')
plt.ylabel('Density')

# Display the plot
plt.show()
```
