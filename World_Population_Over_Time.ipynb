```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv("world-population.csv")

data.head()
```

```python
# Create a line chart

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population'], marker='o')
plt.title('World Population Over Years (Line Chart)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
# Create a step chart

plt.figure(figsize=(10, 6))
plt.step(data['Year'], data['Population'], where='post', marker='o')
plt.title('World Population Over Years (Step Chart)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```
